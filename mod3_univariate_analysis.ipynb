```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "/content/HepatitisCdata.csv"
data = pd.read_csv(file_path)

print("Dataset Preview:")
print(data.head())

# Summary statistics
print("\nSummary Statistics for Numerical Columns:")
print(data.describe())

# Univariate analysis for numerical features
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

for col in numerical_columns:
    print(f"\nUnivariate Analysis for {col}:")
    print(f"Skewness: {data[col].skew()}")
    print(f"Kurtosis: {data[col].kurtosis()}")

    # histogram
    plt.figure(figsize=(8, 4))
    sns.histplot(data[col], kde=True, color='blue')
    plt.title(f"Histogram of {col}")
    plt.xlabel(col)
    plt.ylabel("Frequency")
    plt.grid()
    plt.show()

    # Boxplot for outliers
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[col], color='red')
    plt.title(f"Boxplot of {col}")
    plt.grid()
    plt.show()

# Univariate analysis for categorical features
categorical_columns = data.select_dtypes(include=['object', 'category']).columns

for col in categorical_columns:
    print(f"\nUnivariate Analysis for {col}:")
    print(data[col].value_counts())

    # Bar plot
    plt.figure(figsize=(8, 4))
    sns.countplot(y=data[col], palette="viridis")
    plt.title(f"Count Plot of {col}")
    plt.xlabel("Count")
    plt.ylabel(col)
    plt.grid()
    plt.show()
```
